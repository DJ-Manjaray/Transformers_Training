** AutoBackbone lets you use pretrained models as backbones and get feature maps as outputs from different stages of the models.**

```python
from transformers import AutoImageProcessor, AutoBackbone
import torch
from PIL import Image
import requests
```

```python

url = "http://images.cocodataset.org/val2017/000000039769.jpg" # load an image from coco datasets
image = Image.open(requests.get(url, stream=True).raw) # open image
processor = AutoImageProcessor.from_pretrained("microsoft/swin-tiny-patch4-window7-224")  # image processor
model = AutoBackbone.from_pretrained("microsoft/swin-tiny-patch4-window7-224", out_indices=(0,)) # img-classifer swin from Microsoft
```

```python
inputs = processor(image, return_tensors="pt") # we are using torch here.

outputs = model(**inputs)

feature_maps = outputs.feature_maps

list(feature_maps[-1].shape)
```
